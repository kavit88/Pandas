## Unit-3 Feature Creation & Feature Transformation

```python
import pandas as pd
import numpy as np
data = {
    'A':[10,12,12,15,14],
    'B':[3,4,7,1,6]
}
df = pd.DataFrame(data)
df
```

```python
df['C'] = df['A'] + df['B']
df
```

```python
df['D'] = df['C'] + 4
```

```python
df
```

```python
df['Result'] = np.where(df['D']>18,'Pass','Fail')
df
```

```python
df['Grade'] = np.where(df['D']>22,'A',
                       np.where(df['D']>19,'B',
                                np.where(df['D']>17,'C','D')))
df
```

```python
data = {
    'Marks':[20,21,23,24,25,24,20,21,24,25,24,22],
    'Gender':['Male','Female','Male','Female','Male','Female',
             'Male','Female','Male','Female','Male','Female']
}
df=pd.DataFrame(data)
df
```

## One-Hot Encoding

```python
df = pd.get_dummies(df,drop_first=True)
df
```

## Train-Validation Split

```python
x = df['Gender_Male']
y = df['Marks']

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,
                                                 random_state=10)
x_train
```
